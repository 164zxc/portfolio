### 影片類型分析

```python
fig,ax =plt.subplots(figsize=(8,5))
data.groupby('影片類型')['觀看次數'].mean().plot.barh(ax=ax,color=['#008FD5','#FFA07A'],width=0.8)
plt.ylabel(' ')
plt.title('平均觀看次數比較')
```

目前本頻道的流量較高的是影片，shorts沒有特別突出的表現

```python
mean_shorts = (data[data['影片類型'] == 'shorts']['獲得的訂閱人數'].sum()) / data[data['影片類型'] == 'shorts'].shape[0]
mean_movies = (data[data['影片類型'] == '影片']['獲得的訂閱人數'].sum()) / data[data['影片類型'] == '影片'].shape[0]

# 绘制直方图
fig,ax =plt.subplots(figsize=(8,5))
ax.barh(['shorts', '影片'], [mean_shorts, mean_movies],color=['#008FD5','#FFA07A'])
plt.title('每支影片平均新增訂閱數')
```

頻道訂閱來源依舊是影片更高於short，但也有因為shorts讓一些人認識我們

在每次觀看點讚數這邊發現shorts是超過影片的，雖然差距不大，但可以得知我們shorts的影片質量不輸影片
